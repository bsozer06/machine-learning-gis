## Preparing the dataset

```python
# Import Python 3's print function and division
from __future__ import print_function, division

# Import GDAL, NumPy, and matplotlib
from osgeo import gdal, gdal_array
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Tell GDAL to throw Python exceptions, and register all drivers
gdal.UseExceptions()
gdal.AllRegister()
```

```python
# Read in our image and ROI image
img_ds = gdal.Open('./example/LE70220491999322EDC01_stack.gtif', gdal.GA_ReadOnly)
roi_ds = gdal.Open('./example/training_data.gtif', gdal.GA_ReadOnly)

data_type_enum = img_ds.GetRasterBand(1).DataType

img = np.zeros(
    (img_ds.RasterYSize, img_ds.RasterXSize, img_ds.RasterCount),
    gdal_array.GDALTypeCodeToNumericTypeCode(data_type_enum)
)

print('img_ds shape: ', img_ds.RasterXSize, img_ds.RasterYSize, img_ds.RasterCount)
print('roi_ds shape: ',  roi_ds.RasterXSize, roi_ds.RasterYSize, roi_ds.RasterCount)
print('img shape: ', img.shape)
```

```python
for b in range(img.shape[2]):
    img[:, :, b] = img_ds.GetRasterBand(b + 1).ReadAsArray()

roi = roi_ds.GetRasterBand(1).ReadAsArray().astype(np.uint8)
```

```python
# Display them
plt.subplot(121)
plt.imshow(img[:, :, 4], cmap = plt.cm.Greys_r)
plt.title('SWIR1')

plt.subplot(122)
plt.imshow(roi, cmap = plt.cm.Spectral)
plt.title('ROI Training Data')

plt.show()
```

```python
# Find how many non-zero entries we have -- i.e. how many training data samples?
num_samples = (roi > 0).sum()
print('We have {n} non-zero samples'.format(n = num_samples))

# What are our classification labels?
labels = np.unique(roi[roi > 0])
print('The training data include {n} classes: {classes}'.format(n = labels.size, classes = labels))
```

```python
# The matrix containing our features
X = img[roi > 0, :]

# The array containing our labels
y = roi[roi > 0]

print('Our X matrix is sized: {sz}'.format(sz = X.shape))
print('Our y array is sized: {sz}'.format(sz = y.shape))
```

```python
# Mask out clouds, cloud shadows, and snow using Fmask
clear = X[:, 7] <= 1
```

```python
X = X[clear, :7]  # we can ditch the Fmask band now
y = y[clear]

print('After masking, our X matrix is sized: {sz}'.format(sz = X.shape))
print('After masking, our y array is sized: {sz}'.format(sz = y.shape))
```

## Training the Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize our model with 500 trees
rf = RandomForestClassifier(n_estimators=500, oob_score=True)

# Fit our model to training data
rf = rf.fit(X, y)
```

#### Random Forest Out-of-Bag (OOB) Prediction Score

```python
print('Our OOB prediction of accuracy is: {oob}%'.format(oob=rf.oob_score_ * 100))
```

```python
# Specify which spectral bands are more important based on feature importance score
bands = [1, 2, 3, 4, 5, 6, 7]

for b, imp in zip(bands, rf.feature_importances_):
        print('Band {b} importance: {imp}'.format(b=b, imp=imp))

# SWIR1 and the Green bands were the most useful
```

### Predicting the rest of image

```python
# Take our full image by ignoring the Fmask band,
# and reshape into long 2d array (nrow * ncol, nband) for classification

new_shape_size = (img.shape[0] * img.shape[1], img.shape[2] - 1)

img_as_array = img[:, :, :7].reshape(new_shape)
print('Reshaped from {o} to {n}'.format(o = img.shape, n = img_as_array.shape))
```

```python
# Now predict for each pixel
class_prediction = rf.predict(img_as_array)

# Reshape our classification map
class_prediction = class_prediction.reshape(img[:, :, 0].shape)
```

### Visualize

```python
def color_stretch(image, index, minmax=(0, 10000)):
    colors = image[:, :, index].astype(np.float64)
    
    max_val = minmax[1]
    min_val = minmax[0]
    
    # Enforce maximum and minimum values
    colors[colors[:, :, :] > max_val] = max_val
    colors[colors[:, :, :] < min_val] = min_val
    
    # Normalizing 
    for b in range(colors.shape[2]):
        colors[:, :, b] = colors[:, :, b] * 1 / (max_val - min_val)
        
    return colors
```

```python
# First setup a 5-4-3 composite
img543 = color_stretch(img, [4, 3, 2], (0,8000))

# Next setup a colormap for our map
colors = dict((
    (0, (0, 0, 0, 255)),  # Nodata
    (1, (0, 150, 0, 255)),  # Forest
    (2, (0, 0, 255, 255)),  # Water
    (3, (0, 255, 0, 255)),  # Herbaceous
    (4, (160, 82, 45, 255)),  # Barren
    (5, (255, 0, 0, 255))  # Urban
))


# Put 0 - 255 as float 0 - 1
for k in colors:
    v = colors[k]
    _v = [_v / 255.0 for _v in v]
    colors[k] = _v

n = class_prediction.max() 
    
#index_colors = [colors[key] if key in colors else (255, 255, 255, 0) for key in range(1, n + 1)]
index_colors = []
for key in range(1, n + 1):
    if key in colors:
        print('key {n}:'.format(n=key), colors[key])
        index_colors.append(colors[key])
    else:
        index_colors.append((255, 255, 255, 0))
        
cmap = plt.matplotlib.colors.ListedColormap(index_colors, 'Classification', n)
```

```python
# Now show the classmap next to the image
plt.subplot(121)
plt.imshow(img543)

plt.subplot(122)
plt.imshow(class_prediction, cmap=cmap, interpolation='none')

plt.show()
```
